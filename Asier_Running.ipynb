```python
coord_map = {
    "asier": {
        "lat": 43.2630,
        "lon": -2.9350,
        "timezone": "Europe/Madrid",
        "location": "Bilbao"
    },
    "runner_log": {
        "lat": 41.8781,
        "lon": -87.6298,
        "timezone": "America/Chicago",
        "location": "Chicago"
    },
    "runner_garmin": {
        "lat": 34.7304,
        "lon": -86.5861,
        "timezone": "America/Chicago",  # Huntsville también está en Central
        "location": "Huntsville"
    },
    "ganix": {
        "lat": 43.2630,
        "lon": -2.9350,
        "timezone": "Europe/Madrid",
        "location": "Bilbao"
    },
    "ari": {
        "lat": 43.2630,
        "lon": -2.9350,
        "timezone": "Europe/Madrid",
        "location": "Bilbao"
    },
}
```

```python
def get_location(row):
    info = coord_map.get(row["runner"])
    return info["location"] if info else pd.NA
```

```python
import requests

def get_weather_df(lat, lon, start_date, end_date):
    url = "https://archive-api.open-meteo.com/v1/archive"

    params = {
        "latitude": lat,
        "longitude": lon,
        "start_date": start_date,
        "end_date": end_date,
        "hourly": "temperature_2m,wind_speed_10m,relative_humidity_2m,precipitation",
        "timezone": "auto"
    }

    data = requests.get(url, params=params).json()

    wdf = pd.DataFrame({
        "datetime_hour": pd.to_datetime(data["hourly"]["time"]),
        "temp_C": data["hourly"]["temperature_2m"],
        "wind_kmh": data["hourly"]["wind_speed_10m"],
        "humidity": data["hourly"]["relative_humidity_2m"],
        "rain_mm": data["hourly"]["precipitation"],
    })

    return wdf

```

```python
weather_dfs = []

for runner, cfg in coord_map.items():
    sub = runs[runs["runner"] == runner].copy()
    if sub.empty:
        continue

    start = sub["datetime"].min().date().isoformat()
    end = sub["datetime"].max().date().isoformat()

    wdf = get_weather_df(cfg["lat"], cfg["lon"], start, end)

    merged = sub.merge(wdf, on="datetime_hour", how="left")
    weather_dfs.append(merged)
```

```python
runs = pd.concat(weather_dfs, ignore_index=True)
```
