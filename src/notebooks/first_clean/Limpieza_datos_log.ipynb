```python
# lista de columnas que queremos conservar
keep_cols = [
    'Date', 'Title', 'Distance', 'Time',
    'Avg HR', 'Avg Run Cadence', 'Avg Pace',
    'Elev Gain', 'Avg Stride Length',
    'Number of Laps'
]

# por si alguna no existe, hacemos intersección
keep_cols = [c for c in keep_cols if c in log_running.columns]

# nos quedamos solo con esas
log_running = log_running[keep_cols].copy()
```

* Hay que limpiar TIME, algunos vienen en HH:MM:SS y otros solo MM:SS
* Hay que cambiar valores -- de todo el DF a nulos
* Borrar filas de nulos de las columnas que necesito si o si

```python
log_running = log_running.replace("--", np.nan)
```

```python
log_running =log_running.dropna(subset=["Avg HR", "Avg Run Cadence"])
```

```python
log_running["Date"] = pd.to_datetime(log_running["Date"])
log_running["duration_s"] = log_running["Time"].apply(time_convert)
log_running["Avg HR"] = pd.to_numeric(log_running["Avg HR"])
log_running["Avg Run Cadence"] = pd.to_numeric(log_running["Avg Run Cadence"])
log_running["pace_min_km"] = log_running["Avg Pace"].apply(pace_to_float)
log_running["pace_s_km"] = log_running["pace_min_km"] * 60
log_running["pace_s_km"] = log_running["pace_min_km"] * 60
```

```python
log_running["Elev Gain"] = (log_running["Elev Gain"].str.replace(",", ""))
log_running["Elev Gain"] = pd.to_numeric(log_running["Elev Gain"])
log_running["Elev Gain"] = log_running["Elev Gain"] / 10
```

```python
log_running.drop(columns=["Time", "Avg Pace", "pace_min_km"], inplace=True)
```

```python
log_running = log_running.rename(columns=rename_map)
```
