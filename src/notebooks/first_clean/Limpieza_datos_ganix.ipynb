```python
import pandas as pd
import numpy as np
```

```python
ganix_running = pd.read_csv("../../data/actividades_running/ganix_running.csv")
```

```python
ganix_running.head()
```

```python
ganix_running.WorkoutType.unique()
```

Ganix coger Run

```python
ganix_running = ganix_running[ganix_running.WorkoutType == "Run"]
```

```python
ganix_running.WorkoutType.unique()
```

```python
ganix_running = ganix_running.dropna(axis=1, how="all")
```

```python
ganix_running.info()
```

```python
ganix_running.columns
```

```python
# lista de columnas que queremos conservar
keep_cols = [
    'Title',
       'WorkoutDay', 'DistanceInMeters',
       'TimeTotalInHours', 'CadenceAverage',
       'HeartRateAverage',
]

# por si alguna no existe, hacemos intersección
keep_cols = [c for c in keep_cols if c in ganix_running.columns]

# nos quedamos solo con esas
ganix_running = ganix_running[keep_cols].copy()
```

```python
ganix_running["distance_km"] = ganix_running["DistanceInMeters"] / 1000
ganix_running["duration_s"] = ganix_running["TimeTotalInHours"] * 3600
```

```python
ganix_running["date"] = pd.to_datetime(ganix_running["WorkoutDay"])
```

```python
ganix_running.drop(columns=["WorkoutDay", "DistanceInMeters", "TimeTotalInHours"], inplace=True)
```

```python
ganix_running.head()
```

```python
ganix_running.info()
```

Tengo que borrar los nulos 

```python
ganix_running = ganix_running[ganix_running["distance_km"].notna()]
```

```python
rename_map = {
    "Title": "title",
    "distance_km": "distance_km",
    "HeartRateAverage": "avg_hr",
    "CadenceAverage": "avg_cadence",
    "date": "date",
    "duration_s": "duration_s"
}
```

```python
ganix_running = ganix_running.rename(columns=rename_map)
```

```python
ganix_running["runner"] = "ganix"
```

```python
final_cols = [
    "runner", "date", "title",
    "distance_km", "duration_s",
    "avg_hr", "avg_cadence"
]

ganix_running = ganix_running[final_cols]
```
