```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ruta al archivo
file = "../../data/actividades_running/ari_running.json"

with open(file, "r", encoding="utf-8") as f:
    data = json.load(f)

# data es una lista con un solo diccionario
activities_list = data[0]["summarizedActivitiesExport"]

# Esto ya sí es una lista de dicts → DataFrame bien formateado
ari_running = pd.DataFrame(activities_list)
```

```python
ari_running[ari_running["activityType"] == "treadmill_running"]
```

activityType de Ari running, treadmill_running y trail_running pero hay que mirarlo más a fondo

```python
ari_running = ari_running[(ari_running.activityType == "running") | (ari_running.activityType == "treadmill_running") | (ari_running.activityType == "trail_running")]
```

```python
ari_running.activityType.unique()
```

```python
ari_running.info()
```

Había demasiadas columnas y he borrado las que están llenas de nulos, voy a hacerlo con todos

```python
ari_running = ari_running.dropna(axis=1, how="all")
```

```python
ari_running.info()
```

Vamos a hacer limpieza de columnas, hay demasiadas

```python
# lista de columnas que queremos conservar
keep_cols = [
    # Identificación / contexto
    "name", "activityType",
    "startTimeLocal", "locationName",
    
    # Métricas básicas
    "duration", "distance",
    "elevationGain",
    
    # Frecuencia cardiaca
    "avgHr",
    
    # Cadencia / técnica
    "avgDoubleCadence","avgStrideLength",
    
    # GPS
    "startLatitude", "startLongitude",

    "lapCount",
]

# por si alguna no existe, hacemos intersección
keep_cols = [c for c in keep_cols if c in ari_running.columns]

# nos quedamos solo con esas
ari_running = ari_running[keep_cols].copy()
```

Lo exporto ya a un CSV para poder ver mejor que es cada columna

```python
ari_running.to_csv("../../data/actividades_running/ari_running.csv", index=False)
```

```python
ari_running.info()
```

```python
ari_running["date"] = pd.to_datetime(ari_running["startTimeLocal"], unit="ms")
```

```python
ari_running.head()
```

```python
ari_running["duration_s"] = ari_running["duration"] / 1000
ari_running["distance_km"] = ari_running["distance"] / 100000
ari_running["elevation_gain_m"] = ari_running["elevationGain"] / 100
ari_running["avg_stride_length_m"] = ari_running["avgStrideLength"] / 100
ari_running["pace_s_km"] = ari_running["duration_s"] / ari_running["distance_km"]
```

```python
ari_running.head()
```

```python
ari_running.drop(columns=["startTimeLocal", "duration", "distance","elevationGain", "avgStrideLength"], inplace=True)
```
