```python
import pandas as pd
import numpy as np

import sys
import os

# Añadir la carpeta utils al path
sys.path.append(os.path.abspath("../../utils"))

from functions import *
```

```python
asier_running = pd.read_csv("../../data/actividades_running/asier_running.csv")
```

```python
asier_running = asier_running.dropna(axis=1, how="all")
```

```python
# lista de columnas que queremos conservar
keep_cols = [
    'Tipo de actividad', 'Fecha', 'Título', 'Distancia',
       'Tiempo', 'Frecuencia cardiaca media',
       'Cadencia de carrera media',
       'Ritmo medio',
       'Ascenso total', 'Longitud media de zancada',
       'Número de vueltas',
       
]

# por si alguna no existe, hacemos intersección
keep_cols = [c for c in keep_cols if c in asier_running.columns]

# nos quedamos solo con esas
asier_running = asier_running[keep_cols].copy()
```

```python
asier_running["Tipo de actividad"].unique()
```

```python
asier_running = asier_running[asier_running["Tipo de actividad"] == "Carrera"]
```

```python
asier_running.head()
```

```python
asier_running.info()
```

```python
asier_running["date"] = pd.to_datetime(asier_running["Fecha"])
```

```python
asier_running["duration_s"] = asier_running["Tiempo"].apply(time_convert)
```

```python
asier_running["pace_min_km"] = asier_running["Ritmo medio"].apply(pace_to_float)
asier_running["pace_s_km"] = asier_running["pace_min_km"] * 60
```

```python
asier_running["Ascenso total"] = pd.to_numeric(asier_running["Ascenso total"])
```

```python
asier_running.drop(columns=["Tipo de actividad", "Fecha", "Tiempo","Ritmo medio", "pace_min_km"], inplace=True)
```

```python
rename_map = {
    "Título": "title",
    "Distancia": "distance_km",
    "Frecuencia cardiaca media": "avg_hr",
    "Cadencia de carrera media": "avg_cadence",
    "Ascenso total": "elevation_gain_m",
    "Longitud media de zancada": "avg_stride_length_m",
    "Número de vueltas": "laps",
    "date": "date",
    "duration_s": "duration_s",
    "pace_s_km": "pace_s_km"
}
```

```python
asier_running = asier_running.rename(columns=rename_map)
```

```python
asier_running["runner"] = "asier"
```

```python
final_cols = [
    "runner", "date", "title",
    "distance_km", "duration_s", "pace_s_km",
    "avg_hr", "avg_cadence",
    "avg_stride_length_m", "elevation_gain_m", "laps"
]

asier_running = asier_running[final_cols]
```
