```python
asier_running = pd.read_csv("../data/actividades_running_limpias/asier_running.csv")
ari_running = pd.read_csv("../data/actividades_running_limpias/ari_running.csv")
garmin_running = pd.read_csv("../data/actividades_running_limpias/garmin_running.csv")
log_running = pd.read_csv("../data/actividades_running_limpias/log_running.csv")
```

```python
runs = pd.concat(
    [asier_running, ari_running, garmin_running, log_running],
    ignore_index=True
)
```

```python
runs["date"] = pd.to_datetime(runs["date"])
```

```python
runs["time_of_day"] = runs["date"].apply(get_time_of_day)
```

```python
runs[runs["avg_stride_length_m"] > 1.6].head()
```

```python
runs["pace_min_km"] = runs["pace_s_km"] / 60
runs = runs[runs["distance_km"] > 1]
runs = runs[runs["avg_hr"] > 120]
runs = runs[runs["pace_min_km"] < 8.5]
runs = runs[runs["elevation_gain_m"] < 300]
runs = runs[runs["distance_km"] < 25]
runs = runs[runs["avg_stride_length_m"] < 1.8]
```

```python
runs[runs["runner"] == "log"] = runs[runs["pace_min_km"] > 5]
```

```python
runs.info()
```

```python
runs.info()
```

```python
runs = runs[(runs["laps"] - runs["distance_km"]) < 4]
```

Creamos eficiencia, (explicar)

```python
runs["efficiency"] = runs["pace_s_km"] * runs["avg_hr"]
```

Creamos un DF de kms semanales acumulados y eficiencia media semanal

```python
runs["year_week"] = runs["date"].dt.strftime("%G-%V")
```

```python
weekly_km = (
    runs.groupby(["runner", "year_week"])["distance_km"]
      .sum()
      .reset_index()
      .rename(columns={"distance_km": "km_week"})
)
weekly_km
```

```python
weekly_km["km_prev_week"] = (
    weekly_km
    .groupby("runner")["km_week"]
    .shift(1)
)
weekly_km
```

```python
weekly_eff = (
    runs.groupby(["runner", "year_week"])["efficiency"]
      .mean()
      .reset_index()
)
```

```python
weekly_eff["eff_prev"] = weekly_eff.groupby("runner")["efficiency"].shift(1)
weekly_eff["eff_improvement"] = weekly_eff["eff_prev"] - weekly_eff["efficiency"]
```

eff_improvement - **Positivo** más eficiente

```python
weekly_eff
```

Juntamos Km semanales y eficiencia media semanal para la hipotesis 3

```python
weekly = weekly_km.merge(weekly_eff, on=["runner", "year_week"], how="left")
```
